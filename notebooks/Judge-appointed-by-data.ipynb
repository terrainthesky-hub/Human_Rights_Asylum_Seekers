```python
# Government website with 30 appeal immagration judges
url = 'https://www.justice.gov/eoir/board-of-immigration-appeals-bios#WilliamACassidy'

# bs4 gets web html
response = requests.get(url)
soup = Bs4(response.text, 'html.parser')

# Current judges of the appeal's court
bodytext = soup.find('div', class_='bodytext')
#names = []
bio = []
# find all names
for content in bodytext.find_all('p'):
    bio.append(content.text)
# Clean list of information
for i in bio:
    if i == 'Return to top of the page':
        j = bio.index(i)
        bio.pop(j)
        j = None
bio = bio[4:]
```

```python
# Separate bio and name into individual lists, also, clean data
judges_bio = []
judges_names = []
for i in bio:
    # find all bios
    if len(i) > 100:
        # adding bios to bio list
        judges_bio.append(i)
        # checking for names in bios
        if '\n' in i:
            j = i.index('\n')
            # adding the names to list
            judges_names.append(i[:j])
    else:
        # adding the names to list
        judges_names.append(i)
judges_names.pop(-2)
judges_names.pop(-15)
judges_names.pop(-8)
```

```python
def scrapelotsofjudges(u):
    # create space with list
    name = [] 
    bio = []
    appointed_date = []
    administration = []
    department = []
    # loop though each section of text
    for i in range(0,len(u)):
        # catch larger text sections
        while 
        if 135 <= len(u[i]):
            key = "\t\t\t"
            key2 = '\n\n'
            key_end ='\n\n\n'
            y = u[i].find(key) + 3
            x = u[i][y:].find(key) + 3 + y
            w = u[i][x:].find(key) + 3 + x
            d = u[i][w:].find(key2) + 2 + w
            a = u[i][d:].find(key_end)+d
            t = u[i][d:a]
            j = t.find(key)-1
            # first judge add its the outliner in the html
            name.append(t[:j])
            # finding the count of the rest of names in text
            c = count_seqence(t, key)
            f = t[j+4:]
            while c >= 1:
                m = f.find(key)-1
                judge = f[:m]
                name.append(judge)
                # reformat for while loop
                f = f[m+4:]
    return name, bio, appointed_date, administration, department
name, bio, appointed_date, administration, department = scrapelotsofjudges(u)
print(name)
```
