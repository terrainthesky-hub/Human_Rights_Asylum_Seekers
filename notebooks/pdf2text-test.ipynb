```python
import pdftotext

# Load your PDF
with open("test1.pdf", "rb") as f:
    pdf = pdftotext.PDF(f)

# # If it's password-protected
# with open("secure.pdf", "rb") as f:
#     pdf = pdftotext.PDF(f, "secret")

# How many pages?
print("Pages:", len(pdf))

# Iterate over all the pages
for page in pdf:
    print(page)

# # Read some individual pages
# print(pdf[0])
# print(pdf[1])

# Read all the text into one string
print("\n\n".join(pdf))
```
