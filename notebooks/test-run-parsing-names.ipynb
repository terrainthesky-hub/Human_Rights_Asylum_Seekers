```python
import pdftotext


# define your pdf
x = "test2.pdf"

# Load your PDF
with open(x, "rb") as f:
    pdf = pdftotext.PDF(f)

# How many pages?
print("Pages:", len(pdf))

# Iterate over all the pages
for page in pdf:
    
# # Read some individual pages
    document = pdf[0]
    document1 = pdf[1]
    document2 = pdf[2]

# Read all the text into one string
# print("\n\n".join(pdf))
```

```python
import spacy
# loading the large english model
nlp = spacy.load("en_core_web_lg")
```

```python
# First page 
doc = nlp(document)
```

```python
# getting index of the token 
[token.i for token in doc]
```

```python
# boolean value if it contains alphabet 
[token.is_alpha for token in doc]
```

```python
# boolean value to check if the token is a punctuation
[token.is_punct for token in doc]
```

```python
# if token is a number 
[token.like_num for token in doc]
```

# named entities 

```python
for ent in doc.ents:
    # print the entity text and the label 
    print(ent.text, ent.label_)
```

```python
# Printing just the names 
for ent in doc.ents:
    # print the entity text and the label 
    if ent.label_ == "PERSON":
        
        print([ent.text.strip('\n').strip().replace("\n","")])
```

```python
doc_pg2 = nlp(document1)

storage = []
for ent in doc_pg2.ents:
    # print the entity text and the label 
    if ent.label_ == "PERSON":
        if ent.text not in storage:
            storage.append(ent.text.strip().replace("\n",""))

print(storage)

# Having an issue with printing in the format of (last name, first name)
```

# using Matcher 

```python
from spacy.matcher import Matcher 
```

```python
# initialize the matcher 
matcher = Matcher(nlp.vocab)

# Add the pattern to the matcher 

pattern = [{"ENT_TYPE": "PERSON"}, {"ORTH": ","}, {"ENT_TYPE": "PERSON"}]
matcher.add("judge name", None, pattern) # 2nd argument is an optional callback

# call the matcher on the doc 
matches = matcher(doc)
```

```python
# returns a match_id, start ind and end index
matches
```

```python
# iterate over matches
# works just fine but we are missing with the middle name initial 
for _, start, end in matches:
    matched_span = doc[start:end]
    print(matched_span.text)
```
